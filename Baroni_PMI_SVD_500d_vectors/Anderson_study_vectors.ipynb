```python
import pandas as pd

# Load the list of words from your CSV file
stimuli = pd.read_csv('stimuli.csv')
words = stimuli['term'].tolist()

# Initialize a dictionary to store embeddings
embeddings = {word: [] for word in words}

# Read the embeddings file
with open('EN-wform.w.2.ppmi.svd.500.txt', 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.split()
        word = parts[0]
        if word in embeddings:
            # Convert embedding values to float and store them
            embeddings[word] = [float(val) for val in parts[1:]]

# Convert the dictionary to a DataFrame
embeddings_df = pd.DataFrame.from_dict(embeddings, orient='index').transpose()

# Save the DataFrame to a CSV file
#embeddings_df.to_csv('word_embeddings.csv', index=False)
```

### Extract mithell words

```python
# Define the new set of words
new_words = ["see", "hear", "listen", "taste", "smell", "eat", "touch", "rub", "lift", "manipulate",
             "run", "push", "fill", "move", "ride", "say", "fear", "open", "approach", "near",
             "enter", "drive", "wear", "break", "clean"]

# Initialize a dictionary for the new embeddings
new_embeddings = {word: [] for word in new_words}

# Read the embeddings file and extract embeddings for the new words
with open('EN-wform.w.2.ppmi.svd.500.txt', 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.split()
        word = parts[0]
        if word in new_embeddings:
            # Convert embedding values to float and store them
            new_embeddings[word] = [float(val) for val in parts[1:]]

# Convert the new embeddings dictionary to a DataFrame
new_embeddings_df = pd.DataFrame.from_dict(new_embeddings, orient='index').transpose()
```

```python
# Now calculate the correlation matrix between embeddings_df and new_embeddings_df
# First, ensure both DataFrames have the same number of rows (filling missing values if necessary)
embeddings_df = embeddings_df.reindex(range(new_embeddings_df.shape[0])).fillna(0)
```

```python
embeddings_df 
```

```python
# Initialize an empty DataFrame with the desired shape
final_matrix = pd.DataFrame(index=new_embeddings_df.columns, columns=embeddings_df.columns)

# Compute the correlation for each pair of words
for new_word in new_embeddings_df.columns:
    for original_word in embeddings_df.columns:
        # Compute correlation
        correlation = embeddings_df[original_word].corr(new_embeddings_df[new_word])
        # Store the correlation in the final matrix
        final_matrix.loc[new_word, original_word] = correlation

# Display the shape of the final matrix to verify
final_matrix.shape
```

```python
### correlation wiont run because the output are objects -- annoying!
# Convert all columns to numeric
for col in final_matrix.columns:
    final_matrix[col] = pd.to_numeric(final_matrix[col], errors='coerce')

# After conversion, check if there are any NaN values
print(final_matrix.isna().sum())

# Compute the correlation matrix
correlation_matrix = final_matrix.corr()

correlation_matrix.to_csv('anderson_baroni_similarity.csv')

# Display the correlation matrix
print(correlation_matrix)



```
